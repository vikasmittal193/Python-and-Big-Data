```python
### merge or join  - inner, left, right, outer
df1 = pd.DataFrame({'id':['A','B','C'],'val1':[20,21,34]})
df2 = pd.DataFrame({'id':['C','D','E'],'val2':[55,44,11]})
```

```python
df_inner = pd.merge(df1,df2, how = 'inner', on ='id')
df_inner
df_inner.columns = ['id', 'left', 'right']
df_inner
```

```python
df_left = pd.merge(df1,df2, how='outer', on='id')
#df_left.dropna(inplace=True)
df_left
```

```python
#### self join in pandas
data = {'transaction_id': [1, 2, 3, 4, 5], 
        'close_date': ["2012-08-01", "2012-08-02", "2012-08-03", "2012-08-04", "2012-08-04"], 
        'buyer_name': ["Julia", "Joe", "Jake", "Jamie", "Jackie"],
       'seller_name': ["Lara", "Julia", "Barbara", "Emily", "Mason"]
       }
df = pd.DataFrame(data)
df
```

```python
# to check people who are buyer as well as seller
df1df2 = df.merge(right=df, left_on='buyer_name', right_on = 'seller_name')
df1df2
```

```python
df_melt = pd.melt(df, id_vars=['transaction_id'], value_vars=['close_date','buyer_name','seller_name'], var_name='Variable',value_name = 'value')
df_melt
```

```python
df_pivot = df_melt.pivot(index = 'transaction_id',columns='Variable', values='value')
print(df_pivot)
df_pivot = df_pivot.reset_index()
print(df_pivot)
```
